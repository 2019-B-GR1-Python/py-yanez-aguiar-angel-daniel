```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
lista = pd.read_csv(path, sep=';')
```

```python
p = sns.barplot(x='Sexo', y='Numero_asesinatos', hue='Estado Civil', data=lista)
```

```python
data_suicide_mean = lista['Numero_asesinatos'].groupby(lista.Provincia).mean().sort_values(ascending=False)
f,ax = plt.subplots(1,1,figsize=(15,4))
ax = sns.barplot(data_suicide_mean.head(5).index,data_suicide_mean.head(5),palette='coolwarm')
```

```python
p = sns.barplot(x='tipo_muert_matriz', y='Numero_asesinatos', data=lista)
```

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(1, 1, 1)
ax.plot(lista.groupby(['Edad']).Numero_asesinatos.sum(),'ko--')
tick = ax.set_xticks(range(0,80,10))
ax.set_title('Total suicidios de 0 a 80 años')
ax.set_xlabel('Edad')
ax.set_ylabel('Numero_asesinatos')
```

```python
#Top causas de muerte
serie_count_muertes=lista.Presunta_Motivacion.value_counts().head(5)
#serie_count_muertes['OTROS']=lista.shape[0]-serie_count_muertes.sum()
patches,texts=plt.pie(serie_count_muertes,startangle=90,shadow=True)
plt.legend(patches,serie_count_muertes.index,loc="best")
plt.axis('equal')
plt.title("Top muertes")
plt.tight_layout()
plt.show()
```

```python
serie_count_muertes_sin_otros=lista.Tipo_Arma.value_counts().head(5)
atches,texts=plt.pie(serie_count_muertes_sin_otros,startangle=90,shadow=True)
plt.legend(patches,serie_count_muertes_sin_otros.index,loc="best")
plt.axis('equal')
plt.title("Top muertes")
plt.tight_layout()
plt.show()
```
