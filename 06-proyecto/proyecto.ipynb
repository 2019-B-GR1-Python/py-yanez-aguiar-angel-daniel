```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### leer el archivo csv: 

```python
path='C://Users/Angel/Documents/GitHub/py-yanez-aguiar-angel-daniel/06-proyecto/asesinatos.csv'
```

```python
lista = pd.read_csv(path, sep=';')
```

##### Determinar las provincias con mayor indice de homicidios y asesinatos

```python
data_suicide_mean = lista['Numero_asesinatos'].groupby(lista.Provincia).mean().sort_values(ascending=False)
f,ax = plt.subplots(1,1,figsize=(15,4))
ax = sns.barplot(data_suicide_mean.head(5).index,data_suicide_mean.head(5),palette='coolwarm')
```

Análisis: las porvinicias donde mas se registran asesinatos y homicidios son Santa Elena, Esmeraldas, Tungurahua, Loja y Guayas.

##### Analizar el numero de homicidios y asesinatos segun el genero y su estado civil.

```python
p = sns.barplot(x='Sexo', y='Numero_asesinatos', hue='Estado Civil', data=lista)
```

Análisis: Respecto al genero masculino, las mayores victimas son los viudos. En el genero femenino, las mujeres en union libre son las mayores victimas de homicidios y asesinatos

```python
serie_homicidios_sexo=lista.Sexo.value_counts()
total_homicidios=lista.shape[0]
porcentaje_hombres=(serie_homicidios_sexo["MASCULINO"]/total_homicidios)*100
porcentaje_mujeres=(serie_homicidios_sexo["FEMENINO"]/total_homicidios)*100
patches,texts=plt.pie(serie_homicidios_sexo,shadow=True,startangle=90)
plt.pie(serie_homicidios_sexo,shadow=True,labels=serie_homicidios_sexo.index,autopct='%1.1f%%')
plt.legend(patches,serie_homicidios_sexo.index,loc="best")
plt.axis("equal")
plt.title("Homicidios según el sexo")
plt.tight_layout()
numero_homicidios_hombres=serie_homicidios_sexo['MASCULINO']
numero_homicidios_mujeres=serie_homicidios_sexo['FEMENINO']
#print(f"Número de homicidios a hombres: {numero_homicidios_hombres}")
#print(f"Porcentaje de homicidios a hombres: {porcentaje_hombres}%")
#print(f"Número de homicidios a mujeres: {numero_homicidios_mujeres}")
#print(f"Porcentaje homicidios a mujeres: {porcentaje_mujeres}%")
plt.show()
```

Análisis: Durante el año 2014 las mayores victimas fueron las personas de genero masculino.

### Comprobar el tipo de muerte que mas se ha llevado a cabo durante este año.

```python
p = sns.barplot(x='tipo_muert_matriz', y='Numero_asesinatos', data=lista)
```

Analisis: Los femicidios se han vuelto el tipo de asesinato que a cobrado mas victimas durante el año 2014, despues lo sigue las muertes por sicariato.

### Determinar la edad con mayores victimas de homicidios y asesinatos.

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(1, 1, 1)
ax.plot(lista.groupby(['Edad']).Numero_asesinatos.sum(),'ko--')
tick = ax.set_xticks(range(0,80,10))
ax.set_title('Total suicidios de 0 a 80 años')
ax.set_xlabel('Edad')
ax.set_ylabel('Numero_asesinatos')
```

Analisis: Se puede evidenciar que las victimas estan entre los 20 a 35 años de edad, este rango de edad es donde existe el mayor numero de asesinatos.

### Analizar la motivacion para cometer los delitos de homicidios y asesinatos.

```python
#Top causas de muerte
serie_count_muertes=lista.Presunta_Motivacion.value_counts().head(5)
#serie_count_muertes['OTROS']=lista.shape[0]-serie_count_muertes.sum()
patches,texts=plt.pie(serie_count_muertes,startangle=90,shadow=True)
plt.legend(patches,serie_count_muertes.index,loc="best")
plt.axis('equal')
plt.title("Top muertes")
plt.tight_layout()
plt.show()
```

Analisis: Se puede evidenciar que las riñas son la principal causa de asesinatos, a este se le suma las muertes por amenazas y el robo a personas contribuye a la suma de victimas.

### Determinar el tipo de arma que se utiliza con mayor frecuencia en estos delitos.

```python
serie_count_muertes_sin_otros=lista.Tipo_Arma.value_counts().head(5)
atches,texts=plt.pie(serie_count_muertes_sin_otros,startangle=90,shadow=True)
plt.legend(patches,serie_count_muertes_sin_otros.index,loc="best")
plt.axis('equal')
plt.title("Top muertes")
plt.tight_layout()
plt.show()
```

Analisis: como es de conocimiento publico y por difusion en medios de comunicacion, la pistola, revolver y cuchillo es el amra que mas se utiliza para cometer asesinatos y homicidios.

```python
serie_count_muertes_sin_otros=lista.Hora_Infraccion.value_counts().head(5)
#atches,texts=plt.pie(serie_count_muertes_sin_otros,startangle=90,shadow=True)
plt.pie(serie_count_muertes_sin_otros,shadow=True,labels=serie_count_muertes_sin_otros.index,autopct='%1.1f%%')
plt.legend(patches,serie_count_muertes_sin_otros.index,loc="best")
plt.axis('equal')
plt.title("Top muertes")
plt.tight_layout()
plt.show()
```

Analisis: las horas con mayor numero de victimas son: 0:00 es decir a la madrugada, 20:30 y 22:00.

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(1, 1, 1)
ax.plot(lista.groupby(['Zona']).Numero_asesinatos.sum())
#tick = ax.set_xticks(range(0,80,10))
ax.set_title('Asesinatos por zona')
ax.set_xlabel('Zona')
ax.set_ylabel('Numero_asesinatos')
```
