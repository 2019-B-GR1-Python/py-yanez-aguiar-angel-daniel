```python
import os
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt
```

```python
mapache = misc.face()
mapache_parts = np.vsplit(mapache,4)
piezas_index = np.random.permutation(4)
mapache_desordenado = np.concatenate(list(map(lambda x: mapache_parts[x],piezas_index)))
plt.imshow(mapache_desordenado)
plt.show()

def menu():
    os.system('clear') # NOTA para windows tienes que cambiar clear por cls
    print ("Selecciona una opcin")
    print ("\t1 - armar rompecabezas")
    print ("\t2 - volver a generar rompecabezas")
    print ("\t3 - salir")

while True:
     # Mostramos el menu
    menu()
    
    # solicituamos una opcin al usuario
    opcionMenu = input("inserta un numero valor >> ")
    
    if opcionMenu=="1":
        mover_posicion_a = input("ingrese la posicion de la pieza que desea mover >> ")
        mover_posicion_b = input("ingrese la posicion en donde va a insertar la pieza >> ")
        pieza_temporal = piezas_index[int(mover_posicion_a)]
        piezas_index[int(mover_posicion_a)] = piezas_index[int(mover_posicion_b)]
        piezas_index[int(mover_posicion_b)] = pieza_temporal
        mapache_desordenado = np.concatenate(list(map(lambda x: mapache_parts[x],piezas_index)))
        plt.imshow(mapache_desordenado)
        plt.show()
    
    elif opcionMenu=="2":
        piezas_index = np.random.permutation(4)
        mapache_desordenado = np.concatenate(list(map(lambda x: mapache_parts[x],piezas_index)))
        plt.imshow(mapache_desordenado)
        plt.show()
        
    elif opcionMenu=="3":
        break
    else:
        print ("")
        input("No has pulsado ninguna opcin correcta...\npulsa una tecla para continuar")
```
