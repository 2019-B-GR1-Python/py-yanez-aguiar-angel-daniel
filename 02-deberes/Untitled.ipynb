```python
import os

def menu():
    os.system('clear') # NOTA para windows tienes que cambiar clear por cls
    print ("Selecciona una opción")
    print ("\t1 - ingresar concesionario")
    print ("\t2 - ingresar auto")
    print ("\t3 - consultar concesionario")
    print ("\t4 - consultar auto")
    print ("\t5 - modificar concesionario")
    print ("\t6 - modificar auto")
    print ("\t7 - eliminar concesionario")
    print ("\t8 - eliminar auto")
    print ("\t9 - salir")
 
 
while True:
    # Mostramos el menu
    menu()
    
    # solicituamos una opción al usuario
    opcionMenu = input("inserta un numero valor >> ")
    
    if opcionMenu=="1":
        print ("")
        nombre = input("ingrese nombre del concesionario >> ")
        direccion = input("ingrese direccion del concesionario >> ")
        telefono = input("ingrese telefono del concesionario >> ")
        #estado_civil = input("ingrese el nombre del padre >> ")
        #cedula = input("ingrese el nombre del padre >> ")
        
        try:
            path = "./padre.txt"
            archivo_abierto_escritura = open(path,mode="a")
            archivo_abierto_escritura.write(f"{nombre}, ")
            archivo_abierto_escritura.write(f"{direccion}, ")
            archivo_abierto_escritura.write(f"{telefono}\n")
            archivo_abierto_escritura.close()
        except Exception as error:
            print('error')
            
    elif opcionMenu=="2":
        print ("")
        modelo = input("ingrese modelo del auto >> ")
        anio = input("ingrese año del auto >> ")
        color = input("ingrese color del auto >> ")
        try:
            path = "./hijo.txt"
            archivo_abierto_escritura = open(path,mode="a")
            archivo_abierto_escritura.write(f"{modelo}, ")
            archivo_abierto_escritura.write(f"{anio}, ")
            archivo_abierto_escritura.write(f"{color}\n")
            archivo_abierto_escritura.close()
        except Exception as error:
            print('error')
    elif opcionMenu=="3":
        print ("")
        try:
            path = "./padre.txt"
            archivo_abierto = open(path)
            contenido = archivo_abierto.readlines()
            for linea in contenido:
                print(linea)
            archivo_abierto.close()
        except Exception as error:
            print('error')
    elif opcionMenu=="4":
        print ("")
        try:
            path = "./hijo.txt"
            archivo_abierto = open(path)
            contenido = archivo_abierto.readlines()
            for linea in contenido:
                print(linea)
            archivo_abierto.close()
        except Exception as error:
            print('error')
    elif opcionMenu=="5":
        print ("")
        try:
            path = "./padre.txt"
            archivo_abierto = open(path)
            contenido = archivo_abierto.readlines()
            archivo_abierto.close()
            archivo_abierto = open(path,mode="w")
            
            for linea in contenido:
                print(linea)
            linea_eliminar = input("ingrese el concesionario a modificar >> ")
            linea_agregar = input("ingrese el concesionario modificado >> ")
            for linea in contenido:
                if linea != f"{linea_eliminar}\n":
                    archivo_abierto.write(f"{linea}")
                else:
                    archivo_abierto.write(f"{linea_agregar}\n")
            archivo_abierto.close()
        except Exception as error:
            print('error')
    elif opcionMenu=="6":
        print ("")
        try:
            path = "./hijo.txt"
            archivo_abierto = open(path)
            contenido = archivo_abierto.readlines()
            archivo_abierto.close()
            archivo_abierto = open(path,mode="w")
            
            for linea in contenido:
                print(linea)
            linea_eliminar = input("ingrese el auto a modificar >> ")
            linea_agregar = input("ingrese el auto modificado >> ")
            for linea in contenido:
                if linea != f"{linea_eliminar}\n":
                    archivo_abierto.write(f"{linea}")
                else:
                    archivo_abierto.write(f"{linea_agregar}\n")
            archivo_abierto.close()
        except Exception as error:
            print('error')
    elif opcionMenu=="7":
        print ("")
        try:
            path = "./padre.txt"
            archivo_abierto = open(path)
            contenido = archivo_abierto.readlines()
            archivo_abierto.close()
            archivo_abierto = open(path,mode="w")
            
            for linea in contenido:
                print(linea)
            linea_eliminar = input("ingrese el concesionario a eliminar >> ")
            for linea in contenido:
                if linea != f"{linea_eliminar}\n":
                    archivo_abierto.write(f"{linea}")
            archivo_abierto.close()
        except Exception as error:
            print('error')
    elif opcionMenu=="8":
        print ("")
        try:
            path = "./hijo.txt"
            archivo_abierto = open(path)
            contenido = archivo_abierto.readlines()
            archivo_abierto.close()
            archivo_abierto = open(path,mode="w")
            
            for linea in contenido:
                print(linea)
            linea_eliminar = input("ingrese el auto a eliminar >> ")
            for linea in contenido:
                if linea != f"{linea_eliminar}\n":
                    archivo_abierto.write(f"{linea}")
            archivo_abierto.close()
        except Exception as error:
            print('error')
    elif opcionMenu=="9":
        break
    else:
        print ("")
        input("No has pulsado ninguna opción correcta...\npulsa una tecla para continuar")
```
