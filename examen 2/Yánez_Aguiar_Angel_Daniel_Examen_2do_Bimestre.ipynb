```python
#7) Crear un Dataframe con 10 registros y 6 columnas y Ordenar el dataframe 1 vez por cada columna, ascendente y descendente
df6 = pd.DataFrame(np.random.randint(1, 10, 60).reshape(-1,6), columns=list('ABCDEF'))
df6
```

```python
def fun1(condicion,df,col): #recibe true para ascendente y false para descendente
    return(df.sort_values(col,ascending=condicion))
```

```python
print(fun1(True,df6,"A"))
print("\n")
print(fun1(True,df6,"B"))
print("\n")
print(fun1(True,df6,"C"))
print("\n")
print(fun1(True,df6,"D"))
print("\n")
print(fun1(True,df6,"E"))
print("\n")
print(fun1(True,df6,"F"))
```

```python
#14) Crear un Dataframe con 10 registros y 3 columnas, A B C, llenas de n√∫meros randomicos del 1 al 10 enteros. Crear una nueva columna con el calculo por fila (A * B ) / C
df13 = pd.DataFrame(np.random.randint(1, 10, 30).reshape(-1,3), columns=list('ABC'))
df13
```

```python
campo_calculado=[]
for i in df13.index:
    calculo=(df13["A"][i]*df13["B"][i])/df13["C"][i]
    campo_calculado.append(calculo)

df13['D'] = campo_calculado
df13
```
