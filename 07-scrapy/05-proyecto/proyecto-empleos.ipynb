##### Determinar las provincias con mayor indice de homicidios y asesinatos

```python
plt.figure(figsize=(10,8))
plt.title('Oferta de trabajo por jornada')
lista['jornada'].value_counts()[:10].plot(kind='bar')
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.title('Oferta de trabajo por localizacion')
lista['localizacion'].value_counts()[:10].plot(kind='bar')
plt.show()
```

Análisis: las porvinicias donde mas se registran asesinatos y homicidios son Santa Elena, Esmeraldas, Tungurahua, Loja y Guayas.

```python
plt.figure(figsize=(10,8))
plt.title('Oferta de trabajo por empresa')
lista['empresa'].value_counts()[:10].plot(kind='bar')
plt.show()
```

```python
serie_count_muertes_sin_otros=lista.contrato[lista.localizacion==' Quito, Pichincha'].value_counts().head(5)
#atches,texts=plt.pie(serie_count_muertes_sin_otros,startangle=90,shadow=True)
plt.pie(serie_count_muertes_sin_otros,shadow=True,labels=serie_count_muertes_sin_otros.index,autopct='%1.1f%%')
plt.legend(patches,serie_count_muertes_sin_otros.index,loc="best")
plt.axis('equal')
plt.title("oferta de trabajo por contrato en Quito, Pichincha")
plt.tight_layout()
plt.show()
```
